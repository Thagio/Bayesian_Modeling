```python
# ライブラリの読み込み
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import pymc3 as pm

%matplotlib inline
```

## 3.2 離散確率分布の学習と予測

```python
n = 1000 # 試行回数
a, b = 1,1
beta_trials = stats.beta.rvs(a,b, size=n)
#mean, var, skew, kurt = stats.beta.stats(a, b, moments='mvsk')
```

```python
# ベータ分布を可視化 -> 一様分布になることを確認
plt.hist(beta_trials, bins='auto')
plt.show()
```

```python
# 疑似データ
trials = stats.bernoulli.rvs(0.5, size=n)
```

```python
plt.hist(trials, bins='auto')
plt.show()
```

```python
# pymc3によるモデル化
with pm.Model() as model:b
    mu = pm.Beta("beta",alpha=1,beta=1)
    y = pm.Bernoulli('y',p=mu,observed=trials)
```

```python
# モデルのグラフィカル表示（3.5から追加された機能）
pm.model_to_graphviz(model)
```

```python
# MCMCの実行
with model:
    trace = pm.sample(2000, tune=500, step=pm.NUTS())b
```

```python
# 収束チェック
pm.traceplot(trace)
```

```python
# 結果の出力
pm.summary(trace)
```

```python
# 事後分布の表示
pm.plot_posterior(trace)
```
